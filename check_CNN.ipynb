# Load

```python
import tensorflow as tf
from google.colab import drive

# 1. Mount Google Drive
drive.mount('/content/drive')

# 2. Path to your model (change this to match your file location in Drive)
model_path = '/content/drive/MyDrive/research/my_model.keras'

# 3. Load the saved model
model = tf.keras.models.load_model(model_path)

# 4. Check the model structure
model.summary()
```

# Test

```python
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

class_names = ['Cataract', 'Diabetic Retinopathy', 'Glaucoma', 'Normal']

img_path = '/content/100_left.jpeg'

img = tf.keras.utils.load_img(img_path, target_size=(256, 256))
img_array = tf.keras.utils.img_to_array(img)

img_array = np.expand_dims(img_array, axis=0).astype(np.float32)

pred = model.predict(img_array)
predicted_class = np.argmax(pred[0])
confidence = np.max(pred[0])

print(f"Tahmin edilen sınıf: {class_names[predicted_class]} ({confidence*100:.2f}%)")

plt.imshow(img)
plt.title(f"Tahmin: {class_names[predicted_class]}")
plt.axis('off')
plt.show()
```
